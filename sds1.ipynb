```python
def batsman_score_value():
    # total run percentile and percentage
    top_batsman_run=df.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False)

    batsman_run=top_batsman_run[batsman_name]

    top_batsman_size=top_batsman_run.size # total number of batsman

    mask=top_batsman_run.values>batsman_run
    better_batsman=np.sum(mask) #no of true i.e people above the threashold

    percentile_run=(top_batsman_size-better_batsman)/top_batsman_size*100
    percentile_run

    percentage_run=(top_batsman_run[batsman_name]/top_batsman_run[0])*100
    percentage_run

    #mean_run_score=stats.tmean([percentage_run,percentile_run])
    mean_run_score=percentile_run
    #print(percentage_run,percentile_run)
    mean_run_score

    #bests batsman strikerate percentage and percentile
    top_batsman_ball=df.groupby("batsman")["batsman_runs"].count()
    top_batsman_ball=top_batsman_ball[top_batsman_ball>190]
    top_batsman_runrate=(top_batsman_run/top_batsman_ball).sort_values(ascending=False)

    batsman_runrate=top_batsman_runrate[batsman_name]

    top_batsman_size=top_batsman_runrate.size # total number of batsman

    mask=top_batsman_runrate.values>batsman_runrate
    better_batsman=np.sum(mask) #no of true i.e people above the threashold

    percentile_runrate=(top_batsman_size-better_batsman)/top_batsman_size*100
    percentile_runrate

    percentage_runrate=(top_batsman_runrate[batsman_name]/top_batsman_runrate[0])*100
    percentage_runrate

    #mean_runrate_score=stats.tmean([percentage_runrate,percentile_runrate])
    mean_runrate_score=percentile_runrate
    #print(percentage,percentile)
    mean_runrate_score


    batsman_score=stats.tmean([mean_run_score,mean_runrate_score])
    return batsman_score
```

```python
def bowler_score_value():
    
    # wicket score
    
    top_bowler=df.groupby("bowler")["dismissal_kind"].count().sort_values(ascending=False)
    bowler_wicket=top_bowler[bowler_name]

    top_bowler_size=top_bowler.size # total number of batsman

    mask=top_bowler.values>bowler_wicket
    better_bowler=np.sum(mask) #no of true i.e people above the threashold

    percentile=(top_bowler_size-better_bowler)/top_bowler_size*100
    percentile

    percentage=(top_bowler[bowler_name]/top_bowler[0])*100
    percentage

    #mean_wicket_score=stats.tmean([percentage,percentile])
    mean_wicket_score=percentile
    

    # economy rate

    bowler_ball=df["bowler"].value_counts()
    #bowler_ball=total_ball[total_ball.values>170]

    bowler_runs=df.groupby("bowler")["total_runs"].sum().sort_values(ascending=False)

    economy_rate=(bowler_runs/bowler_ball).dropna().sort_values()

    economy_rate_bowler=economy_rate[bowler_name]

    mean_economy_score=(economy_rate>economy_rate_bowler).sum()/economy_rate.shape[0]*100


    bowler_score=stats.tmean([mean_wicket_score,mean_economy_score])
    return bowler_score
```

```python
batsman_name=input()
bowler_name=input()

try:
    if(bowler_score_value(bowler_name)>batsman_score_value(batsman_name)):
        print(" Bowler, "+bowler_name+" performs well")
    else:
        print("Batsman, "+ batsman_name +" perfoms well")
except:
    print("Invalid")

```
